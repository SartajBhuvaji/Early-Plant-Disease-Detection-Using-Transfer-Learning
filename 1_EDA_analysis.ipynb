```python
import statistics
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/metadata.csv')
```

```python
df.head()
```

```python
#Sorting the DF
df.sort_values('PlantVillage', inplace=True)
print(df.head())
```

```python
#Creating a new DF that would help in choosing the plants for our project

info_dict ={"Type":['Mean','Std dev','Classes']} 
```

```python
apples = df.iloc[0:4,0:4]
plt.barh(apples.PlantVillage, apples.Count)
print("Mean :", apples['Count'].mean())
print("Std dev :", statistics.stdev(apples.Count))

info_dict["Apple"]= [apples['Count'].mean(), statistics.stdev(apples.Count),4]
info_dict
```

```python
blueberry = df.iloc[4:5,0:4]
plt.barh(blueberry.PlantVillage, blueberry.Count)
#Single row hence no mean and std dev
```

```python
cherries = df.iloc[5:7,0:4]
plt.barh(cherries.PlantVillage, cherries.Count)
print("Mean :", cherries['Count'].mean())
print("Std dev :", statistics.stdev(cherries.Count))

info_dict["Cherries"] = [cherries['Count'].mean(), statistics.stdev(cherries.Count),2]
```

```python
corns = df.iloc[7:11,0:4]
plt.barh(corns.PlantVillage, corns.Count)
print("Mean :", corns['Count'].mean())
print("Std dev :", statistics.stdev(corns.Count))

info_dict["Corn"] = [corns['Count'].mean(), statistics.stdev(corns.Count), 4]
info_dict
```

```python
grapes = df.iloc[11:15,0:4]
plt.barh(grapes.PlantVillage, grapes.Count)
print("Mean :", grapes['Count'].mean())
print("Std dev :", statistics.stdev(grapes.Count))

info_dict["Grapes"] = [grapes['Count'].mean(), statistics.stdev(grapes.Count), 4]
```

```python
orange = df.iloc[15:16,0:4]
plt.barh(orange.PlantVillage, orange.Count)
#Single row hence no mean and std dev
```

```python
peppers = df.iloc[20:22,0:4]
plt.barh(peppers.PlantVillage, peppers.Count)
print("Mean :", peppers['Count'].mean())
print("Std dev :", statistics.stdev(peppers.Count))

info_dict["Peppers"] = [peppers['Count'].mean(), statistics.stdev(peppers.Count), 2]
```

```python
potatoes = df.iloc[22:28,0:4]
plt.barh(potatoes.PlantVillage, potatoes.Count)
print("Mean :", potatoes['Count'].mean())
print("Std dev :", statistics.stdev(potatoes.Count))

info_dict["Potatoes"] = [potatoes['Count'].mean(), statistics.stdev(potatoes.Count), 3]
info_dict
```

```python
tomatoes = df.iloc[33:,0:4]
plt.barh(tomatoes.PlantVillage, tomatoes.Count)
print("Mean :", tomatoes['Count'].mean())
print("Std dev :", statistics.stdev(tomatoes.Count))

info_dict["Tomatoes"] = [tomatoes['Count'].mean(), statistics.stdev(tomatoes.Count), 20]
info_dict
```
